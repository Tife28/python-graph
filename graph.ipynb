```python
from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
x=np.arange(0, math.pi*2, 0.05)
y=np.sin(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, label=r'$f(x)=\frac{\sin(x)}{x}$')
ax.set_title('sine wave')
ax.set_xlabel('angle')
ax.set_ylabel('sine')
```
